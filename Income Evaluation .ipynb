### Change all the categorical columns into numeric using dummy variables (One Hot Encoding) & store it in a new Dataframe

```python
pd.set_option('display.max_columns', None)
```

```python
w=pd.get_dummies(df)
```

```python
w.head()
```

### Split the dataset into input and target variables

```python
X=w.drop(columns=' income')
y=w[' income']
```

```python
X.head()
```

```python
y
```

### Carry out feature scaling with Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
xscaler=scaler.fit_transform(X)
```

```python
X=pd.DataFrame(xscaler,columns=X.columns)
```

```python
X.head()
```

### Split the dataset into training & Testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 100)
```

### Check the shape of X_train and X_test

```python
X_train.shape
```

```python
X_test.shape
```

### Create Gaussian Naive Bayes model

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

### Train the model

```python
model.fit(X_train,y_train)
```

### Check the score of our model

```python
model.score(X_train,y_train)
```

### Make predictions with X_test

```python
y_pred=model.predict(X_test)
```

### Check the accuracy score of our prediction

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_pred)
```

### Create confusion matrix

```python
metrics.confusion_matrix(y_test,y_pred)
```

### Visualize confusion matrix

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_pred),annot=True,fmt='d')
```

### Create classification report

```python
print(metrics.classification_report(y_test,y_pred))
```
